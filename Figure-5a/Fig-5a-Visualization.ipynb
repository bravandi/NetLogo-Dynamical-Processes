```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np
import os
import random
import time
from tqdm import tqdm
from itertools import combinations
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib as mpl
import datetime
import math
from scipy.stats import norm, kurtosis, mode
cmap = plt.get_cmap('tab10')
from scipy.stats import describe
import seaborn as sns
from matplotlib.pyplot import figure
import pickle
```

```python
k_avg_l = [4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
instances = range(100)
```

```python
rn_kavg = []

for k_avg in tqdm(k_avg_l):
    temp = []
    for j in instances:
        try:
            ct = np.load('res/5a-ER-k_avg' + str(k_avg) + 'instance-' + str(j) + '.pickle_endtime.npy')
            temp.append(describe(ct)[4])
        except:
            continue 
    print(len(temp))
    rn_kavg.append(sum(i > 1 for i in temp))
```

```python
figure(figsize=(12,10), dpi=300)
plt.plot(k_avg_l, rn_kavg, marker="^", markersize=10)
plt.xlabel('Average Degree')
plt.ylabel('Percentage of ER Graphs with High Skewness')
plt.savefig('Figure-5a-new.png')
plt.show()
```
